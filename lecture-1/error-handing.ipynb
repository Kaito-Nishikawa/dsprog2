```python
try:
    resault = x /y
except ZeroDivisionError as e:
    resault = None
    print("エラーが発生しました")
    print("0で割ることはできません。結果をnoneに設定します。")

print("結果",resault)

try:
    resault = x / z
except ZeroDivisionError as e:
    resault = None
    print("エラーが発生しました")
    print("変数が見つかりません。結果をnoneに設定します。")

print("結果",resault)
```

```python
## エラーが発生しなかった場合のみ処理をする場合
```

```python
z=3
try:
    resault = x /y
except ZeroDivisionError as e:
    resault = None
    print("エラーが発生しました")
    print("0で割ることはできません。結果をnoneに設定します。")

print("結果",resault)

try:
    resault = x / z
except ZeroDivisionError as e:
    resault = None
    print("エラーが発生しました")
    print("変数が見つかりません。結果をnoneに設定します。")

else:
    #tryブロック内でエラーが発生しなかった場合のみ実行される
    print("エラーは発生しませんでした。結果を表示します。")

print("結果",resault)
```

```python
##エラーの発生に関わらず実行したい処理がある場合
```

```python
try:
    resault = x /y
except ZeroDivisionError as e:
    resault = None
    print("エラーが発生しました")
    print("0で割ることはできません。結果をnoneに設定します。")

print("結果",resault)

try:
    resault = x / z
except ZeroDivisionError as e:
    resault = None
    print("エラーが発生しました")
    print("変数が見つかりません。結果をnoneに設定します。")

else:
    #tryブロック内でエラーが発生しなかった場合のみ実行される
    print("エラーは発生しませんでした。結果を表示します。")
finally:
    #エラーの発生に関わらず実行される
    print("処理が終了しました。")
print("結果",resault)
```

```python
## システム終了以外全てハンドリングする場合
```

```python
x=1
y=0
z=3
try:
    resault = x / y
except Exception as e:
    resault = None
    print("エラーが発生しました",e)
    print("エラーオブジェクト:",type(e))
    print("0で割ることはできません。結果をnoneに設定します。")

else:
    #tryブロック内でエラーが発生しなかった場合のみ実行される
    print("エラーは発生しませんでした。結果を表示します。")
finally:
    #エラーの発生に関わらず実行される
    print("処理が終了しました。")
print("結果",resault)
```

## webスクレイピングでのエラー処理

```python
import requests
```

```python
url = "https://www.musashino-u.ac.jp/invalid-page"#存在しないURLURL
response = requests.get(url)
print(f"ステータスコード：{response.status_code}")

#ステータスコードが200以外の場合に例外を発生させる
response.raise_for_status()
```

```python
url = "https://www.musashino-u.ac.jp/invalid-page"#存在しないURLURL

try:
    response = requests.get(url)
    response.raise_for_status()
except requests.exceptions.RequestException as e:
    print("ウェブページの取得中にエラーが発生しました:", e)
except requests.exceptions.RequestException as e:
    print("ウェブページの取得中に予期しないエラーが発生しました:", e)
else:
    print("ウェブページが正常に取得されました。")
    print(response.text)
finally:
    print("ウェブスクレイピング処理が終了しました。")


```
